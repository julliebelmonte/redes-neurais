#### Redes Neurais Convolucionais (CNN)

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
# vamos comparar os diversos tipos de threshold

img = cv2.imread('threshold2.png')

img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# img_blur = cv2.medianBlur(img_gray, 5)

img_blur = img_gray

_, th_binary = cv2.threshold(img_blur, 127, 255, cv2.THRESH_BINARY)

th_gaussian = cv2.adaptiveThreshold(img_blur, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 31, 2)

th_mean = cv2.adaptiveThreshold(img_blur, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 31, 2)

th, th_otsu = cv2.threshold(img_blur, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# th_mean = cv2.medianBlur(th_mean, 3)
# th_gaussian = cv2.medianBlur(th_gaussian, 3)


titles = ['Original', 'Global (127)', f'Otsu ({th})', 'Adapt. Mdia', 'Adapt. Gaussiano']
images = [img_blur, th_binary, th_otsu, th_mean, th_gaussian]

plt.figure(figsize=(8,8))
for i in range(5):
    plt.subplot(2, 3, i+1)
    plt.imshow(images[i], cmap = 'gray')
    plt.title(titles[i])
    plt.axis('off')
plt.show()
```
