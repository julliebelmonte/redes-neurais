#### PDI
Processamento Digital de Imagem consiste em aplicar operações matemáticas para alterar os valores dos pixels de uma ou mais imagens. 
Uma imagem digital pode ser considerada como sendo uma matriz cujos índices de linhas e colunas identificam um ponto na imagem (pixel)

As cores primárias de pigmentos são definidas como cores que absorvem uma cor primária da luz e reflete as outras duas:
Magenta = R + B;
Cyan = G + B;
Yellow = R + G.

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
def exibe_image(img, gray=True):
    
    if gray:
        plt.imshow(img, cmap = 'gray')
    else:
        plt.imshow(img)
    
    plt.axis('off')
    plt.show()
```

```python
imagem = cv2.imread("lena_color.png")

cv2.imshow("Lena", imagem)

cv2.waitKey(0)

cv2.destroyWindow("Lena")
```

```python
# exibir canais de imagem

blue, green, red = cv2.split(imagem)

blue_image = np.zeros(imagem.shape, dtype = "uint8")
blue_image[:,:,0]  = blue

green_imagem = np.zeros(imagem.shape, dtype = "uint8")
green_imagem[:,:,1]  = green

red_image = np.zeros(imagem.shape, dtype = "uint8")
red_image[:,:,2]  = red

rgb_image = np.concatenate([blue_image, green_imagem, red_image], axis = 1)

plt.figure(figsize = (8,8))
plt.imshow(rgb_image)
plt.axis('off')
plt.show()
```

#### Limiarização
Esse processo é utilizado para separar os níveis de cinza de uma imagem, de tal forma que grupos de pixels de intensidade próxima sejam agrupados. Esse processo aumenta o detalhe de imagens na escala de cinza. Vamos utilizar a imagem da modelo sueca Lena.

```python
img = cv2.imread('lena_color.png')

img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

exibe_image(img_gray)
```

```python
th = 127
img_binary = np.where(img_gray > th, 255, 0)

exibe_image(img_binary)
```

```python
#limiarização global

_, img_binary  = cv2.threshold(img_gray, 127, 255, cv2.THRESH_BINARY)

exibe_image(img_binary)
```

```python
_, img_binary_inv = cv2.threshold(img_gray, 127, 255, cv2.THRESH_BINARY_INV)

exibe_image(img_binary_inv)
```

```python
sum = img_binary_inv - img_binary 

exibe_image(sum)
```

```python
#podemos utilizar um algoritmo para decidir qual o valor ideal de binarização. 

th, img_binary = cv2.threshold(img_gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

print(th)
exibe_image(img_binary)
```

```python
_, img_binary = cv2.threshold(img_gray, 117, 255, cv2.THRESH_BINARY)

exibe_image(img_binary)
```
